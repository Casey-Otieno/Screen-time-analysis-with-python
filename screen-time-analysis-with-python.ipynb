Screen Time Analysis is used to find out how much time one spends on what kind of application and websites on their device. It is used to analyse and create a report on which applications and websites are used and for how long. To perform this task, we will use this [dataset](http:///kaggle/input/analyzing-screen-time) and it contains:
* Date
* Usage of Applications
* Number of notifications from applications
* Number of times apps opened

1. ***Importing necessary python libraries and the dataset***

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)




data=pd.read_csv('/kaggle/input/analyzing-screen-time/Screentime - App Details.csv')
data.head()
```

2. ***Descriptive statistics of the data***

```python
data.describe()
```

3. ***Amount of usage of applications according to date***

```python
fig= px.bar(data, x="Date", y="Usage", color= "App", title="Usage of Applications")
fig.show()
```

4. ***Number of notifications from the apps***

```python
fig= px.bar(data, x="Date", y="Notifications", color= "App", title="Number of Notifications")

fig.show()
```

5. ***Number of times the apps are opened***


```python
fig= px.bar(data, x="Date", y="Times opened", color= "App", title="Number of times opened")

fig.show()
```

6. ***Relationship between the number of notifications and amount odf usage***

```python
fig= px.scatter(data, x="Notifications", y="Usage", 
                trendline="ols", title='Relationship between number of Notifications and Usage')

fig.show()
```

From above, we can conclude that there is linear relationship between the number of notifications and the amount of usage.
